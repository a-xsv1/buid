OpenCV allows reading different types of images (JPG, PNG, etc). You can load grayscale images 
colour images or you can also load images with Alpha channel\
It uses the imread function as follows:

```python
imagePath = "number_zero.jpg"

# Read image in Grayscale format
testImage = cv2.imread(imagePath,0)
print(testImage)
```

cv2.IMREAD_GRAYSCALE or 0: Loads image in grayscale mode\
cv2.IMREAD_COLOR or 1: Loads a color image. Any transparency of image will be neglected. It is the default flag.
cv2.IMREAD_UNCHANGED or -1: Loads image as such including alpha channel.\

```python
imagePath = "Pycharm.jpg"
```

```python
# Read image in Grayscale format
testImage = cv2.imread(imagePath,0)
plt.imshow(testImage)
```

```python
# Read image in Color format
testImage = cv2.imread(imagePath,1)
plt.imshow(testImage)
```

```python
# Read image with alpha channel
testImage = cv2.imread(imagePath,-1)
plt.imshow(testImage)
```

```python
imagePath = "RGB_illumination.jpg"
# Read RGB Image 
testImage = cv2.imread(imagePath)
plt.subplot(121);plt.imshow(testImage);plt.title("Incorrect Image Loading");
plt.subplot(122);plt.imshow(testImage[:,:,::-1]);plt.title("Matplotlib Loading");
```

```python
# Show the channels
plt.figure(figsize=[20,5])

plt.subplot(141);plt.imshow(testImage[:,:,0]);plt.title("Blue Channel");
plt.subplot(142);plt.imshow(testImage[:,:,1]);plt.title("Green Channel");
plt.subplot(143);plt.imshow(testImage[:,:,2]);plt.title("Red Channel");
# Show the merged output
plt.subplot(144);plt.imshow(testImage[:,:,::-1]);plt.title("Merged Output");
```
